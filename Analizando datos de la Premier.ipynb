Pregunta 2. Gráficas. En cada caso interpretar los resultados -Grafique en un histograma las edades de los jugadores. Un histograma usa bins o límites de un
valor numérico, este bins debe ser igual a 5

```python
import matplotlib.pyplot as plt

plt.hist(df['age'].dropna(), bins=5, edgecolor='black')
plt.title('Distribución de edades de los jugadores')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

Conocer la correlación entre variables edad(‘age’) y la variable minutos_jugados
(minutes_played_away)

```python
import matplotlib.pyplot as plt

corr = df['age'].corr(df['minutes_played_away'])
print(f'Correlación de Pearson: {corr:.3f}')
plt.scatter(df['age'], df['minutes_played_away'], alpha=0.5)
plt.title('Relación entre Edad y Minutos jugados (visitante)')
plt.xlabel('Edad')
plt.ylabel('Minutos jugados (visitante)')
plt.show()
```

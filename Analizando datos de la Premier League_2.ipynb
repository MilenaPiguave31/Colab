Conocer la correlación entre variables edad(‘age’) y la variable minutos_jugados
(minutes_played_away)

```python
import matplotlib.pyplot as plt

corr = df['age'].corr(df['minutes_played_away'])
print(f'Correlación de Pearson: {corr:.3f}')
plt.scatter(df['age'], df['minutes_played_away'], alpha=0.5)
plt.title('Relación entre Edad y Minutos jugados (visitante)')
plt.xlabel('Edad')
plt.ylabel('Minutos jugados (visitante)')
plt.show()
```

Objetivo: Representar patrones en tus datos - Histogramas

Representa la cantidad de minutos jugados en un Histograma

```python
import matplotlib.pyplot as plt
df_filtrado = df.dropna(subset=['minutes_played_away', 'position'])
posiciones = df_filtrado['position'].unique()
for pos in posiciones:
    data = df_filtrado[df_filtrado['position'] == pos]['minutes_played_away']

    plt.hist(data, bins=10, edgecolor='black')
    plt.title(f'Minutos jugados (visitante) - {pos}')
    plt.xlabel('Minutos jugados')
    plt.ylabel('Frecuencia')
    plt.show()
```

Representa los mejores jugadores de la Liga en un Histograma

```python
top_players = df[['full_name', 'goals_involved_per_90_overall']].dropna()

top_players = top_players.sort_values(
    by='goals_involved_per_90_overall', ascending=False
).head(10)

print(top_players)
```

```python
import matplotlib.pyplot as plt
df_filtrado = df[['full_name', 'goals_involved_per_90_overall']].dropna()
top_players = df_filtrado.sort_values(
    by='goals_involved_per_90_overall', ascending=False
).head(10)
plt.figure()
plt.barh(top_players['full_name'], top_players['goals_involved_per_90_overall'])
plt.gca().invert_yaxis()
plt.title('Top 10 mejores jugadores (Goles + Asistencias por 90 min)')
plt.xlabel('Goles involucrados por 90 min')
plt.ylabel('Jugador')
plt.show()
```

```python
import matplotlib.pyplot as plt
df_filtrado = df[['full_name', 'goals_involved_per_90_overall']].dropna()
top_players = df_filtrado.sort_values(
    by='goals_involved_per_90_overall', ascending=False
).head(3)
plt.figure()
plt.barh(top_players['full_name'], top_players['goals_involved_per_90_overall'])
plt.gca().invert_yaxis()
plt.title('Top 3 mejores jugadores (Goles + Asistencias por 90 min)')
plt.xlabel('Goles involucrados por 90 min')
plt.ylabel('Jugador')
plt.show()
```

```python
import matplotlib.pyplot as plt
df_filtrado = df[['full_name', 'minutes_played_home']].dropna()

top_home = df_filtrado.sort_values(
    by='minutes_played_home', ascending=False
).head(10)
plt.figure()
plt.barh(top_home['full_name'], top_home['minutes_played_home'])
plt.gca().invert_yaxis()
plt.title('Top 10 jugadores con más minutos jugados en casa')
plt.xlabel('Minutos jugados en casa')
plt.ylabel('Jugador')

plt.show()
```
